```python
# Dependencies and Setup
import pandas as pd
import os
# File to Load (Remember to Change These)
#school_data_to_load = "C:\\Users\\GMyer\\Desktop\\ClassRepo\\git\\pandas_challenge\\PyCitySchools\\schools_complete.csv"
school_data_to_load = os.path.join('PyCitySchools', 'schools_complete.csv')
#student_data_to_load = "C:\\Users\\GMyer\\Desktop\\ClassRepo\\git\\pandas_challenge\\PyCitySchools\\students_complete.csv"
student_data_to_load = os.path.join('PyCitySchools', 'students_complete.csv')
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#total schools
total_schools = school_data_complete["school_name"].nunique()
```

```python
#total students
total_students = school_data_complete["Student ID"].nunique()
```

```python
#total budget* only consider budget per school, not per student
#refernce original csv data?
#groupby
total_budget = school_data["budget"].sum()
```

```python
#average math score
avg_math = school_data_complete["math_score"].mean()
```

```python
#average read score
avg_read = school_data_complete["reading_score"].mean()
```

```python
#passing math score
pass_math = school_data_complete[school_data_complete["math_score"] > 69.99]
#percent passing math
perc_pass_math = (len(pass_math) / total_students )
```

```python
#passing read score
pass_read = school_data_complete[school_data_complete["reading_score"] > 69.99]
#percent passing reading
perc_pass_read = (len(pass_read) / total_students )
```

```python
# passing read and math
pass_both = school_data_complete.loc[(school_data_complete["reading_score"] > 69.9) & (school_data_complete["math_score"] > 69.99)]
pass_both
# percent passed both math and reading(overall passing)
perc_pass_both = (len(pass_both) / total_students )
```

```python
# District Summary Dataframe
# list of dicts or dict of lists?
dis_sum =pd.DataFrame({"Total Schools": [total_schools],"Total Students": [total_students], "Total Budget": [total_budget], "Avg. Math Score": [avg_math], "Avg. Reading Score": [avg_read], "% Passing Math": [perc_pass_math], "% Passing Reading": [perc_pass_read], "% Overall Passing": [perc_pass_both]})
#format at very end
#dataframe.style?
```

```python
#formatting to match starter example
dis_sum["Total Budget"] = dis_sum["Total Budget"].map("${:,.2f}".format)
dis_sum["% Passing Math"] = (dis_sum["% Passing Math"]*100).map("{:.6f}%".format)
dis_sum["% Passing Reading"] = (dis_sum["% Passing Reading"]*100).map("{:.6f}%".format)
dis_sum["% Overall Passing"] = (dis_sum["% Overall Passing"]*100).map("{:.6f}%".format)
dis_sum["Total Students"] = dis_sum["Total Students"].map("{:,}".format)
dis_sum
```

```python
school_summary_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_summary_data.head()
```

```python
#maybe merge then select columns that matter
# School Name

# School Type
school_type_col = school_data.set_index(["school_name"])["type"]
# Total Students
school_students_col = school_summary_data.groupby("school_name").mean()["size"]
#school_students_col = school_summary_data["school_name"].value_counts()
# Total School Budget
school_budget_col = school_summary_data.groupby("school_name").mean()["budget"]
# Per Student Budget
per_stu_budg = school_budget_col / school_students_col
# Average Math Score
avg_math_sch = school_summary_data.groupby("school_name").mean()["math_score"]
# Average Reading Score
avg_read_sch = school_summary_data.groupby("school_name").mean()["reading_score"]
# % Passing Math
    #passing math
num_psmath_sch = school_data_complete[school_data_complete["math_score"] > 69.99].groupby(["school_name"]).count()["student_name"]
    # % pass math
perc_psmath_sch = num_psmath_sch / school_students_col
# % Passing Reading
    #passing reading
num_psread_sch = school_data_complete[school_data_complete["reading_score"] > 69.99].groupby(["school_name"]).count()["student_name"]
    # % passing reading
perc_psread_sch = num_psread_sch / school_students_col
# % Overall Passing (The percentage of students that passed math and reading.)
    #passing both
num_psboth_sch = school_data_complete.loc[(school_data_complete["reading_score"] > 69.9) & (school_data_complete["math_score"] > 69.99)].groupby(["school_name"]).count()["student_name"]
    # % passing both
perc_psboth_sch = num_psboth_sch / school_students_col
school_sum =pd.DataFrame({"School Type": school_type_col,"Total Students": school_students_col,"Total Budget": school_budget_col,"Per Student Budget": per_stu_budg, "Average Math Score": avg_math_sch, "Average Reading Score": avg_read_sch, "% Passing Math": perc_psmath_sch, "% Passing Reading": perc_psread_sch, "% Overall Passing": perc_psboth_sch})
school_sum.head()
```

```python
#formatting for units and such
school_sum_frmt = school_sum.copy()
school_sum_frmt["Total Budget"] = school_sum_frmt["Total Budget"].map("${:,.2f}".format)
school_sum_frmt["Per Student Budget"] = school_sum_frmt["Per Student Budget"].map("${:,.2f}".format)
school_sum_frmt["% Passing Math"] = (school_sum_frmt["% Passing Math"]*100).map("{:.6f}%".format)
school_sum_frmt["% Passing Reading"] = (school_sum_frmt["% Passing Reading"]*100).map("{:.6f}%".format)
school_sum_frmt["% Overall Passing"] = (school_sum_frmt["% Overall Passing"]*100).map("{:.6f}%".format)
school_sum_frmt["Total Students"] = school_sum_frmt["Total Students"].map("{:,}".format)
school_sum_frmt
```

```python
#Math Scores By Grade
#not using .filter/lambda functions due to TypeErrors with series.
#creating new df for each grade instead, then applying avg grade agg.
nine_grade = school_summary_data[school_summary_data["grade"] == "9th"].groupby("school_name").mean()["math_score"]
ten_grade = school_summary_data[school_summary_data["grade"] == "10th"].groupby("school_name").mean()["math_score"]
eleven_grade = school_summary_data[school_summary_data["grade"] == "11th"].groupby("school_name").mean()["math_score"]
twelve_grade = school_summary_data[school_summary_data["grade"] == "12th"].groupby("school_name").mean()["math_score"]
#put into one df
math_by_grade =pd.DataFrame({"9th":nine_grade, "10th": ten_grade, "11th": eleven_grade, "12th": twelve_grade})
math_by_grade
```

```python
#Reading Scores By Grade
#Same as above, just for reading instead of math
nine_grade_r = school_summary_data[school_summary_data["grade"] == "9th"].groupby("school_name").mean()["reading_score"]
ten_grade_r = school_summary_data[school_summary_data["grade"] == "10th"].groupby("school_name").mean()["reading_score"]
eleven_grade_r = school_summary_data[school_summary_data["grade"] == "11th"].groupby("school_name").mean()["reading_score"]
twelve_grade_r = school_summary_data[school_summary_data["grade"] == "12th"].groupby("school_name").mean()["reading_score"]
#put into one df
read_by_grade =pd.DataFrame({"9th":nine_grade_r, "10th": ten_grade_r, "11th": eleven_grade_r, "12th": twelve_grade_r})
read_by_grade
```

```python
#Scores by School Spending
#binning
bins = [0, 585, 610, 635, 660]
labels = ["<$585", "$585-610", "$610-635", "$635-660"]
school_sum["Spending Range per Student"] = pd.cut(per_stu_budg,bins,labels=labels, include_lowest=True)
school_spend_mean_df = school_sum.groupby("Spending Range per Student").mean()
school_spend_mean_df["Total Budget"] = school_spend_mean_df["Total Budget"].map("${:,.2f}".format)
school_spend_mean_df["Per Student Budget"] = school_spend_mean_df["Per Student Budget"].map("${:,.2f}".format)
school_spend_mean_df["% Passing Math"] = (school_spend_mean_df["% Passing Math"]*100).map("{:.6f}%".format)
school_spend_mean_df["% Passing Reading"] = (school_spend_mean_df["% Passing Reading"]*100).map("{:.6f}%".format)
school_spend_mean_df["% Overall Passing"] = (school_spend_mean_df["% Overall Passing"]*100).map("{:.6f}%".format)
school_spend_mean_df["Total Students"] = school_spend_mean_df["Total Students"].map("{:,}".format)
school_spend_mean_df
```

```python
#Scores by School Size
#binning again
bins_s = [0, 1300, 2600, 3900, 5300]
labels_s = ["Small <1300", "Medium 1300-2600", "Large 2600-3900", "Extra Large 3900-5300"]

school_sum["School Size By Enrollment"] = pd.cut(school_students_col,bins=bins_s,labels=labels_s, include_lowest=True)
school_size_mean_df = school_sum.groupby("School Size By Enrollment").mean()
school_size_mean_df["Total Budget"] = school_size_mean_df["Total Budget"].map("${:,.2f}".format)
school_size_mean_df["Per Student Budget"] = school_size_mean_df["Per Student Budget"].map("${:,.2f}".format)
school_size_mean_df["% Passing Math"] = (school_size_mean_df["% Passing Math"]*100).map("{:.6f}%".format)
school_size_mean_df["% Passing Reading"] = (school_size_mean_df["% Passing Reading"]*100).map("{:.6f}%".format)
school_size_mean_df["% Overall Passing"] = (school_size_mean_df["% Overall Passing"]*100).map("{:.6f}%".format)
school_size_mean_df["Total Students"] = school_size_mean_df["Total Students"].map("{:,}".format)
school_size_mean_df
```
